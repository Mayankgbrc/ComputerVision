```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
im = np.zeros((100, 100))

# For a square white in black
im[20:30, 20:30] = 255
plt.imshow(im, cmap = 'gray')

# For digonal
for i in range(len(im)):
    im[i, i] = 255
plt.imshow(im, cmap = 'gray')

# For hollow rect
im[10:20, 70:80] = 255
im[11:19, 71:79] = 0
plt.imshow(im, cmap = 'gray')


# Top bottom edge
im[70, 10:20] = 255
im[80, 10:20] = 255

# Left and right edge
im[70: 80, 10] = 255
im[70: 81, 20] = 255
plt.imshow(im, cmap = 'gray')
```

```python
# Create a pure red image
img = np.zeros((100, 100, 3), dtype = np.uint8)

img[:, :, 0] = 255
img[:, :, 1] = 0
img[:, :, 2] = 0

plt.imshow(img)
```

```python
# Grayscale image to 3 channel (RGB)

img_gray = np.zeros((100, 100), dtype = np.uint8)
img_gray[10:20, 10:20] = 235
img_rgb = np.stack([img_gray]  * 3, axis = -1)
plt.imshow(img_rgb)
```

```python
# Grayscale image to 3 channel (RGB)

img1 = np.zeros((100, 100), dtype = np.uint8) # Red
img2 = np.zeros((100, 100), dtype = np.uint8) # Green
img3 = np.zeros((100, 100), dtype = np.uint8) # Blue
img1[10:20, 10:20] = 255
img2[10:20, 16:26] = 255
img3[16:26, 13:23] = 255
img_rgb = np.stack([img1, img2, img3], axis = -1)
plt.imshow(img_rgb)

# Yellow part is mix of R and G
# Pink is mix of R and B
# Cyan is mix of G and B
# White is mix of R, G, B
```

```python
img_gray = (img_rgb[:, :, 0] * 0.2989) + (img_rgb[:, :, 1] * 0.5870) + (img_rgb[:, :, 2] * 0.1140)
plt.imshow(img_gray, cmap = "gray")
# White, Yello, Cyan, Green, Pink, Red, Blue
```

```python
# Create an RGB image (Red-Green Gradient)
image = np.zeros((256, 256, 3), dtype=np.uint8)
image[:, :, 0] = np.linspace(0, 255, 256)  # Red gradient
image[:, :, 1] = np.linspace(255, 0, 256)  # Green gradient
image[:, :, 2] = 200  # Fixed Blue value

# Custom Grayscale Formula (More Blue, Less Green)
gray_custom = 0.4 * image[:, :, 0] + 0.3 * image[:, :, 1] + 0.3 * image[:, :, 2]
gray_custom = gray_custom.astype(np.uint8)

# Display images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original RGB Image")

plt.subplot(1, 2, 2)
plt.imshow(gray_custom, cmap='gray')
plt.title("Custom Grayscale Image")

plt.show()
```

```python
# Thresholding

# Create a grayscale gradient image
gray_image = np.linspace(0, 255, 256).astype(np.uint8)
gray_image = np.tile(gray_image, (256, 1))  # Repeat for 256 rows

plt.subplot(1,2,1)
plt.imshow(gray_image, cmap = "gray")

threshold = 58
threshold_image = np.where(gray_image > threshold, 255, 0)

plt.subplot(1,2,2)
plt.imshow(threshold_image, cmap = "gray")
```
